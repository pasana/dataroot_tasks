## Start preprocessing

```python
import numpy as np
import nltk
from nltk.metrics import *
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
def removePunctuation(text):
    return re.sub(u'[^а-яА-Я0-9_ ]', ' ', text.lower()).strip()
```

```python
stop_words = stopwords.words('russian')
print "We have %d stop words" % len(stop_words)
```

```python
#lowcase and puncuation
desc=map(lambda x: re.sub(u'[^а-яА-Я0-9_ ]', ' ', x.lower()), desc)
```

```python
desc_tokens = map(lambda x: word_tokenize(x), desc)
```

```python
#stop words
desc_tokens = [filter(lambda x: x not in stop_words, d) for d in desc_tokens]
```

```python
print ' '.join(desc_tokens[5])
```

```python
all_desc_text = '\n'.join(map(lambda x: ' '.join(x), desc_tokens))
```

```python
with open('all_desc_text_no_stop_words.txt', 'w+') as f:
    f.write(all_desc_text.encode('utf-8'))
```
